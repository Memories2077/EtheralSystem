```python
#!npm install -g cloudflared
#!cloudflared tunnel --url http://localhost:11434
!curl -L https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64 -o cloudflared
!chmod +x cloudflared

# 2. Chạy Tunnel với Token từ .env
import os
import subprocess
import threading
from dotenv import load_dotenv

# Load environment variables từ .env
load_dotenv()

# Lấy TUNNEL_TOKEN từ environment variable
TUNNEL_TOKEN = os.getenv("TUNNEL_TOKEN")

if not TUNNEL_TOKEN:
    raise ValueError("TUNNEL_TOKEN not found in .env file. Please add TUNNEL_TOKEN=your_token to .env")

def run_tunnel():
    # Sử dụng token để kết nối vào tunnel cố định đã tạo
    cmd = ["./cloudflared", "tunnel", "--no-autoupdate", "run", "--token", str(TUNNEL_TOKEN)]
    subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE)

threading.Thread(target=run_tunnel, daemon=True).start()
print("Tunnel cố định đang được khởi tạo...")
```
